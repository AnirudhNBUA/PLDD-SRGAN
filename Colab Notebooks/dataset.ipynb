```python
source_path = 'SRGAN/custom_dataset/input/PlantVillage/'
target_path = 'PLDD/custom_dataset/input_64/PlantVillage/'
```

```python
#Create folders inside target path
for folder_name in os.listdir(source_path):
    if not os.path.exists(os.path.join(target_path, folder_name)):
        os.makedirs(os.path.join(target_path, folder_name))
```

```python
for folder_name in os.listdir(source_path):
    for filename in os.listdir(os.path.join(source_path, folder_name)):
        img_path = os.path.join(source_path, folder_name, filename)
        try:
            img = cv2.imread(img_path)
            if img is not None and img.size != 0:
                img = cv2.resize(img, (64, 64))
                target_folder_path = os.path.join(target_path, folder_name)
                cv2.imwrite(os.path.join(target_folder_path, filename), img)
            else:
                print(f"Error reading image: {img_path}")
        except Exception as e:
            print(f"Error processing image: {img_path}. Exception: {str(e)}")
```

```python
from PIL import Image

# set the path to the parent directory containing the subfolders
parent_dir = 'PLDD/custom_dataset/input_64/PlantVillage'

# loop through each subfolder
for folder_name in os.listdir(parent_dir):
    # construct the full path to the subfolder
    folder_path = os.path.join(parent_dir, folder_name)
    
    # check if the item in the parent directory is a folder
    if os.path.isdir(folder_path):
        # count the number of valid image files in the folder
        num_valid_images = 0
        
        # loop through each file in the folder
        for file_name in os.listdir(folder_path):
            # construct the full path to the file
            file_path = os.path.join(folder_path, file_name)
            
            # check if the file is a valid image
            try:
                img = Image.open(file_path)
                img.verify()  # throws an error if the file is not a valid image
                num_valid_images += 1
            except:
                print('Invalid')
        
        # print the folder name and the number of valid image files
        print(f"{folder_name}: {num_valid_images} valid images")
```
